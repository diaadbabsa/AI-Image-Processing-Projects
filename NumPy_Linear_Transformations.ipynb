```python
import numpy as np
import matplotlib.pyplot as plt

def generate_simple_face(transform_matrix=np.eye(2)):
    # Generate base points
    theta = np.linspace(0, 2*np.pi, 100)
    base_points = np.array([np.cos(theta), np.sin(theta)])

    # Transform points
    transformed_points = transform_matrix @ base_points

    # Extract x and y coordinates
    x, y = transformed_points

    # Eyes
    left_eye_base = np.array([-0.3, 0.3])
    right_eye_base = np.array([0.3, 0.3])
    left_eye = transform_matrix @ left_eye_base
    right_eye = transform_matrix @ right_eye_base

    # Mouth (simple curve)
    mouth_base_x = np.linspace(-0.5, 0.5, 100)
    mouth_base_y = -0.1 * np.sin(np.pi * np.linspace(-1, 1, 100)) - 0.3
    mouth_base = np.vstack((mouth_base_x, mouth_base_y))
    mouth_transformed = transform_matrix @ mouth_base

    # Plot the face
    plt.figure(figsize=(8, 8))
    plt.plot(x, y, 'b')
    plt.plot(left_eye[0], left_eye[1], 'bo', markersize=10)
    plt.plot(right_eye[0], right_eye[1], 'bo', markersize=10)
    plt.plot(mouth_transformed[0], mouth_transformed[1], 'r')
    plt.axis('equal')
    plt.title('Transformed Face')
    plt.show()

# Example usage with identity matrix (no transformation)
generate_simple_face()

# Scaling matrix
scale_factor = 1.5
scaling_matrix = np.array([[scale_factor, 0],
                           [0, scale_factor]])
generate_simple_face(scaling_matrix)

# Rotation matrix (45 degrees)
theta = np.pi/4  # 45 degrees in radians
rotation_matrix = np.array([[np.cos(theta), -np.sin(theta)],
                            [np.sin(theta), np.cos(theta)]])
generate_simple_face(rotation_matrix)

# Shear matrix (shear along x-axis)
shear_factor = 1
shear_matrix = np.array([[1, shear_factor],
                         [0, 1]])
generate_simple_face(shear_matrix)

# Singular (non-invertible) matrix
singular_matrix = np.array([[1, 1],
                            [1, 1]])
generate_simple_face(singular_matrix)

# Combination of shear and rotation
combined_matrix = shear_matrix @ rotation_matrix
generate_simple_face(combined_matrix)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def find_line_parameters(point1, point2):
    """
    Find the line parameters for the line passing through two points.

    Parameters:
    point1, point2: tuples or lists containing (x, y) coordinates

    Returns:
    a: slope of the line
    b: y-intercept of the line
    """
    x1, y1 = point1
    x2, y2 = point2

    # Calculate a (slope)
    a = (y2 - y1) / (x2 - x1)

    # Calculate b (y-intercept)
    b = y1 - a * x1

    return a, b

def plot_points_and_line(point1, point2, a, b):
    """
    Plot two points and a line defined by y = ax + b.

    Parameters:
    point1, point2: tuples or lists containing (x, y) coordinates
    a: slope of the line
    b: y-intercept of the line
    """
    # Extract x and y coordinates
    x_coords = [point1[0], point2[0]]
    y_coords = [point1[1], point2[1]]

    # Create line
    x_line = np.array([min(x_coords) - 1, max(x_coords) + 1])
    y_line = a * x_line + b

    # Plot
    plt.figure(figsize=(10, 6))
    plt.scatter(x_coords, y_coords, color='red', s=50, label='Points')
    plt.plot(x_line, y_line, color='blue', label=f'y = {a:.2f}x + {b:.2f}')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Points and Line')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
point1 = (4, 7)
point2 = (2, 3)

# Find the line parameters
a, b = find_line_parameters(point1, point2)

# Plot the points and the line
plot_points_and_line(point1, point2, a, b)
```
