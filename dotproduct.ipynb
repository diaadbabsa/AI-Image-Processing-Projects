```python
import numpy as np

# Question: What is the dot product operation, and how is it implemented in NumPy?

# Define the vectors
vector1 = np.array([1, 2, 3])
vector2 = np.array([4, 5, 6])

# Calculate dot product using NumPy
dot_product_numpy = np.dot(vector1, vector2)

print(f"Dot product using NumPy: {dot_product_numpy}")
```

```python
import numpy as np

# Implement cosine similarity using the dot product
def cosine_similarity(v1, v2):
    dot_product = np.dot(v1, v2)
    norm_v1 = np.linalg.norm(v1)
    norm_v2 = np.linalg.norm(v2)
    return dot_product / (norm_v1 * norm_v2)

# Example vectors
vector_a = np.array([1, 2, 3])
vector_b = np.array([4, 5, 6])

similarity = cosine_similarity(vector_a, vector_b)
print(f"Cosine similarity: {similarity}")
```

```python
# Cell 3: Word Embedding Example with Advanced Model

!pip install sentence-transformers

import numpy as np
from sentence_transformers import SentenceTransformer

# Load a more advanced pre-trained model
model = SentenceTransformer('all-MiniLM-L6-v2')

def get_sentence_embedding(sentence):
    return model.encode(sentence)

def cosine_similarity(v1, v2):
    return np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))

def dot_product_similarity(v1, v2):
    return np.dot(v1, v2)

# Function to compare sentences
def compare_sentences(sentence1, sentence2):
    embedding1 = get_sentence_embedding(sentence1)
    embedding2 = get_sentence_embedding(sentence2)

    cosine_sim = cosine_similarity(embedding1, embedding2)
    dot_sim = dot_product_similarity(embedding1, embedding2)

    print(f"Sentence 1: '{sentence1}'")
    print(f"Sentence 2: '{sentence2}'")
    print(f"Cosine similarity: {cosine_sim:.4f}")
    print(f"Dot product similarity: {dot_sim:.4f}")
    print()

# Example usage
compare_sentences("I love programming", "I enjoy coding")
compare_sentences("dogs are better than cats", "cat is funny!")
compare_sentences("I love pizza", "I hate vegetables")
compare_sentences("im a doctor", "sell some phones!")

print("Comparing sentences with similar meanings but different words:")
compare_sentences("He is very happy", "He feels great joy")

print("Comparing sentences with the same words in different orders:")
compare_sentences("The cat sat on the mat", "On the mat sat the cat")

print("Comparing sentences from different domains or languages:")
compare_sentences("E=mc^2", "La teor√≠a de la relatividad")




# Reflection questions:
print("\nReflection questions:")
print("1. How do cosine similarity and dot product similarity yield identical results in the given examples? What does this reveal about their mathematical properties?")
print("Answer: Cosine similarity and dot product similarity yield identical results when the vectors are normalized. This is because cosine similarity measures the cosine of the angle between two vectors, essentially the dot product of the vectors divided by the product of their magnitudes. When vectors are normalized, their magnitudes are 1, making the cosine similarity equivalent to the dot product.")
print("2. What types of sentences seem to have high similarity? Low similarity?")
print("Answer: Sentences with high similarity typically share similar contexts, meanings, or semantic structures. These sentences often use different words to convey the same idea. Sentences with low similarity have different contexts, meanings, or opposing ideas. Even if they share some common words, the overall meaning can be very different.")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate correlated random points
mean = [0, 0]
cov = [[1, 0.8], [0.8, 1]]  # Covariance matrix with correlation
points = np.random.multivariate_normal(mean, cov, 100)
X = points[:, 0].T
y = points[:, 1].T

# Plot the data points
plt.scatter(X, y)
plt.xlabel('X')
plt.ylabel('y')
plt.title('Correlated Random Points')
plt.show()

# Create the design matrix [X 1]
X_design = np.vstack([X, np.ones(len(X))]).T

# Calculate the pseudo-inverse
X_pseudo_inv = np.linalg.pinv(X_design)

# Calculate the coefficients
coefficients = X_pseudo_inv.dot(y)

# Print the coefficients
print(f"Estimated coefficients: {coefficients.flatten()}")

# Plot the results
plt.scatter(X, y, label='Data points')
plt.plot(X, X_design.dot(coefficients), color='red', label='Linear model')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.title('Linear Regression using Pseudo-Inverse')
plt.show()
```
